#### DIVIDIENDO DATASET EN TRAIN Y TEST
#### GUARDANDO DATASETS CON OUTLIERS (COMPLETO, TRAIN Y TEST)

```python
import pandas as pd

from sklearn.model_selection import train_test_split

predictors = ['minimum_nights', 'number_of_reviews', 'calculated_host_listings_count', 'availability_365', 'room_type_n', 'neighbourhood_group_n']
target = 'price'

x = data_airbnb[predictors]
y = data_airbnb[target]

data_airbnb_outliers = pd.concat ([x,y], axis = 1)
data_airbnb_outliers.to_excel("/workspaces/EDA_LeryBenavides/data/raw/dataset_work/DATASET_AIRBNB-OUTLIERS.xlsx", index = False)

x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.3, random_state = 7)

data_airbnb_outliers_train = pd.concat ([x_train, y_train], axis = 1)
data_airbnb_outliers_train.to_excel("/workspaces/EDA_LeryBenavides/data/raw/dataset_work/DATASET_AIRBNB-OUTLIERS_TRAIN.xlsx", index = False)

data_airbnb_outliers_test = pd.concat ([x_test, y_test], axis = 1)
data_airbnb_outliers_test.to_excel ("/workspaces/EDA_LeryBenavides/data/raw/dataset_work/DATASET_AIRBNB-OUTLIERS_TEST.xlsx", index = False)
```

#### NORMALIZANDO
#### GUARDANDO DATASETS CON OUTLIERS (TRAIN Y TEST)

```python
from sklearn.preprocessing import StandardScaler
from pickle import dump

scaler = StandardScaler()
scaler.fit(x_train)

x_train_norm = scaler.transform (x_train)
x_train_norm = pd.DataFrame (x_train_norm, index = x_train.index, columns = predictors)

x_test_norm = scaler.transform (x_test)
x_test_norm = pd.DataFrame (x_test_norm, index = x_test.index, columns = predictors)

data_airbnb_outliers_train_norm = pd.concat ([x_train_norm, y_train], axis = 1)
data_airbnb_outliers_train_norm.to_excel ("/workspaces/EDA_LeryBenavides/data/raw/dataset_work/DATASET_AIRBNB-OUTLIERS_TRAIN-NORM.xlsx", index = False)

data_airbnb_outliers_test_norm = pd.concat ([x_test_norm, y_test], axis = 1)
data_airbnb_outliers_test_norm.to_excel ("/workspaces/EDA_LeryBenavides/data/raw/dataset_work/DATASET_AIRBNB-OUTLIERS_TEST-NORM.xlsx", index = False)

dump (scaler, open("/workspaces/EDA_LeryBenavides/models/Outliers_normalizer.sav", "wb"))

x_train_norm.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

m_scaler = MinMaxScaler()
m_scaler.fit(x_train)

x_train_scal = m_scaler.transform(x_train)
x_train_scal = pd.DataFrame (x_train_scal, index = x_train.index, columns = predictors)

x_test_scal = m_scaler.transform(x_test)
x_test_scal = pd.DataFrame (x_test_scal, index = x_test.index, columns = predictors)

data_airbnb_outliers_train_scal = pd.concat ([x_train_scal, y_train], axis = 1)
data_airbnb_outliers_train_scal.to_excel ("/workspaces/EDA_LeryBenavides/data/raw/dataset_work/DATASET_AIRBNB-OUTLIERS_TRAIN-SCAL.xlsx", index = False)

data_airbnb_outliers_test_scal = pd.concat ([x_test_scal, y_test], axis = 1)
data_airbnb_outliers_train_scal.to_excel ("/workspaces/EDA_LeryBenavides/data/raw/dataset_work/DATASET_AIRBNB-OUTLIERS_TEST-SCAL.xlsx", index = False)

dump (m_scaler, open("/workspaces/EDA_LeryBenavides/models/Outliers_min-Max.sav", "wb"))

x_train_scal.head()
```

##### CALCULANDO LOS IQR DE LAS VARIABLES 'minimum_nights', 'number_of_reviews', 'calculated_host_listings_count', 'availability_365', 'room_type_n', 'neighbourhood_group_n'

```python
# IQR PARA OUTLIERS DE 'minimum_nights'
min_nights_iqr = descr_1st_var["75%"] - descr_1st_var["25%"]
uplim_1st_var = descr_1st_var["75%"] + 1.5*min_nights_iqr
lowlim_1st_var = descr_1st_var["75%"] - 1.5*min_nights_iqr
if lowlim_1st_var < 0:
    lowlim_1st_var = 0
else: lowlim_1st_var

print("1.- Uplimit de 'min_nights' es:", uplim_1st_var,"y el Lowlimit es:", lowlim_1st_var)

# IQR PARA OUTLIERS DE 'number_of_reviews'
num_rev_iqr = descr_2nd_var["75%"] - descr_2nd_var["25%"]
uplim_2nd_var = descr_2nd_var["75%"] + 1.5*num_rev_iqr
lowlim_2nd_var = descr_2nd_var["25%"] - 1.5*num_rev_iqr
if lowlim_2nd_var < 0:
    lowlim_2nd_var = 0
else: lowlim_2nd_var

print("\n2.- Uplimit de 'num_rev' es:", uplim_2nd_var,"y el Lowlimit es:", lowlim_2nd_var)

# IQR PARA OUTLIERS DE 'calculated_host_listings_count'
calc_host_iqr = descr_3rd_var["75%"] - descr_3rd_var["25%"]
uplim_3rd_var = descr_3rd_var["75%"] + 1.5*calc_host_iqr
lowlim_3rd_var = descr_3rd_var["25%"] - 1.5*calc_host_iqr
if lowlim_3rd_var < 0:
    lowlim_3rd_var = 0
else: lowlim_3rd_var

print("\n3.- Uplimit de 'calc_host' es:", uplim_3rd_var,"y el Lowlimit es:", lowlim_3rd_var)

# IQR PARA OUTLIERS DE 'availability_365'
available_iqr = descr_4th_var["75%"] - descr_4th_var["25%"]
uplim_4th_var = descr_4th_var["75%"] + 1.5*available_iqr
lowlim_4th_var = descr_4th_var["25%"] - 1.5*available_iqr
if lowlim_4th_var < 0:
    lowlim_4th_var = 0
else: lowlim_4th_var

print("\n4.- Uplimit de 'available_365' es:", uplim_4th_var,"y el Lowlimit es:", lowlim_4th_var)

# IQR PARA OUTLIERS DE 'room_type'
room_type_iqr = descr_5th_var["75%"] - descr_5th_var["25%"]
uplim_5th_var = descr_5th_var["75%"] + 1.5*room_type_iqr
lowlim_5th_var = descr_5th_var["25%"] - 1.5*room_type_iqr
if lowlim_5th_var < 0:
    lowlim_5th_var = 0
else: lowlim_5th_var

print("\n5.- Uplimit de 'room_type' es:", uplim_5th_var,"y el Lowlimit es:", lowlim_5th_var)

# IQR PARA OUTLIERS DE 'neighbourhood_group'
neighb_iqr = descr_6th_var["75%"] - descr_6th_var["25%"]
uplim_6th_var = descr_6th_var["75%"] + 1.5*neighb_iqr
lowlim_6th_var = descr_6th_var["25%"] - 1.5*neighb_iqr
if lowlim_6th_var < 0:
    lowlim_6th_var = 0
else: lowlim_6th_var

print("\n6.- Uplimit de 'neighbourhood_group' es:", uplim_6th_var,"y el Lowlimit es:", lowlim_6th_var)
```
